## Creacion de indice de competencia

```python
import pandas as pd
from math import radians, cos, sin, asin, sqrt

def haversine(lon1, lat1, lon2, lat2):
    """
    Calcula la distancia del gran círculo en metros entre dos puntos
    en la tierra (especificados en grados decimales).
    """
    # Convertir grados decimales a radianes
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # Fórmula de Haversine
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    r = 6371000 # Radio de la Tierra en metros.
    return c * r

def calcular_indice_competencia(ruta_csv, actividad_objetivo, radio_metros=500):
    """
    Calcula el índice de competencia para una actividad dada dentro de un radio específico.

    Args:
        ruta_csv (str): La ruta al archivo CSV.
        actividad_objetivo (str): El valor en 'nombre_act' para filtrar.
        radio_metros (int): El radio en metros para buscar competidores.

    Returns:
        pandas.DataFrame: Un DataFrame con el índice de competencia añadido,
                          o None si ocurre un error.
    """
    try:
        # Intenta leer con UTF-8, si falla, prueba con 'latin1' (común en datos de LATAM)
        try:
            df = pd.read_csv(ruta_csv)
        except UnicodeDecodeError:
            print("Probando con codificación 'latin1'...")
            df = pd.read_csv(ruta_csv, encoding='latin1')

    except FileNotFoundError:
        print(f"Error: El archivo '{ruta_csv}' no fue encontrado.")
        return None
    except Exception as e:
        print(f"Error al leer el archivo CSV: {e}")
        return None

    # Filtrar por la actividad objetivo
    df_filtrado = df[df['nombre_act'] == actividad_objetivo].copy()

    if df_filtrado.empty:
        print(f"No se encontraron registros para la actividad: '{actividad_objetivo}'")
        return None

    # Asegurar que latitud y longitud sean numéricas y eliminar filas sin ellas
    df_filtrado['latitud'] = pd.to_numeric(df_filtrado['latitud'], errors='coerce')
    df_filtrado['longitud'] = pd.to_numeric(df_filtrado['longitud'], errors='coerce')
    df_filtrado.dropna(subset=['latitud', 'longitud'], inplace=True)

    print(f"Se encontraron {len(df_filtrado)} tiendas para analizar.")

    indice_competencia = []

    # Iterar sobre cada tienda para calcular su índice
    for index1, row1 in df_filtrado.iterrows():
        competidores = 0
        lat1, lon1 = row1['latitud'], row1['longitud']
        id1 = row1['id'] # Usamos ID para evitar comparar una tienda consigo misma

        # Iterar sobre todas las demás tiendas para comparar distancias
        for index2, row2 in df_filtrado.iterrows():
            if id1 != row2['id']: # No comparar una tienda consigo misma
                lat2, lon2 = row2['latitud'], row2['longitud']
                distancia = haversine(lon1, lat1, lon2, lat2)
                if distancia <= radio_metros:
                    competidores += 1
        indice_competencia.append(competidores)

    # Añadir el índice como una nueva columna al DataFrame filtrado
    df_filtrado['indice_competencia_500m'] = indice_competencia

    return df_filtrado


ruta_del_archivo = 'denue_inegi_19_.csv'
nombre_actividad = 'Comercio al por menor en minisupers'


df_resultados = calcular_indice_competencia(ruta_del_archivo, nombre_actividad)

if df_resultados is not None:
    # Muestra las columnas clave y el nuevo índice
    print("\n--- Resultados del Índice de Competencia ---")
    print(df_resultados[['id', 'nom_estab', 'latitud', 'longitud', 'indice_competencia_500m']].to_string())

    df_resultados.to_csv('resultados_competencia.csv', index=False)
    print("\nResultados guardados en 'resultados_competencia.csv'")
```

```python
df_resultados.to_csv('resultados_competencia.csv', index=False)
print("\nResultados guardados en 'resultados_competencia.csv'")
```
